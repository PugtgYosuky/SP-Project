## Preprocessing dataset

## Analyze past_ variables relation with infringement

```python
past_selected_columns = ['infringed']
for column in infringement_data.columns:
    if column.startswith('past'):
        past_selected_columns.append(column)
```

```python
past_infringement_data = infringement_data[past_selected_columns]
```

```python
past_infringement_data.describe()
```

```python
infringement_data.describe()
```

### Correlation between past credit applications and infringement

```python
past_correlation_all = past_infringement_data.corr()
```

```python
plt.figure()
sns.heatmap(past_correlation_all,
    xticklabels=past_correlation_all.columns,
    yticklabels=past_correlation_all.columns,
)
plt.title('Correlation between past credit applications info and infringements')
```

#### Analyze
Como se pode observar, a correlação entre haver uma infração e a informação de antigos empréstimos é basicamente nula.

### Past infringements averages

```python
past_infringed = past_infringement_data[past_infringement_data.infringed == 1.0]
past_non_infringed = past_infringement_data[past_infringement_data.infringed == 0.0]
```

#### Percentage of each group

```python
total_count = past_infringement_data.shape[0]
infringed_count = past_infringed.shape[0]
non_infringed_count = past_non_infringed.shape[0]
```

```python
percentage_infringed = infringed_count / total_count * 100
percentage_non_infringed = non_infringed_count / total_count * 100
```

```python
print(f'Percentage of infringed data: {percentage_infringed}')
print(f'Percentage of non infringed data: {percentage_non_infringed}')
```

#### Analyze
Como era de esperar, existem muito mais dados sem infrações que com infrações

```python
print('Average of the past loans information when someone infringes the loan')
print(past_infringed.mean())
```

```python
print("Average of the past loans information when someone doesn't infringes the loan")
print(past_non_infringed.mean())
```

#### Analyze

Como se pode observar pelas médias acima calculadas podemos concluir que:
- quem comete uma infração pede em média mais empréstimos do que pessoas que não cometeram infrações
- quem não comete infrações tem em média menos empréstimos recusados e cancelados
- quem não comete uma infração pede em média emprésimos de maior valor em tem mais empréstimos aprovados

No entanto há que ter em conta das diferenças na quantidade de dados que cada categoria e que pessoas que cometeram infrações no passado tem maior probabilidade provavelmente de ter empréstimos aceites após cometerem a infração o que pode justificar ter médias mais baixas nos pontos acima referidos.

## Analysis of the relationship between people's lifestyle and infringements

Para fazer esta análise foram escolhidas algumas variáveis que possam representar o estilo de vida das pessoas

```python
selected_columns_lifestyle_numeric = [
    'infringed',
    'age',
    'annual_income',
    'car_age',
    'days_employed',
    'num_children',
    'num_family_members',
    'region_rating',
    'credit_amount',
    'credit_annuity'

]
selected_columns_lifestyle_categorical = [
    'infringed',
    'contract_type',
    'education',
    'family_status',
    'gender',
    'has_own_car',
    'income_type',
    'occupation_type',
    'has_own_realty',
    'housing_type'
]
```

```python
lifestyle_numeric_data = infringement_data[selected_columns_lifestyle_numeric]
```

```python
lifestyle_categorical_data = infringement_data[selected_columns_lifestyle_categorical]
```

### Analyze numeric data

```python
infringed_lifestyle = lifestyle_numeric_data[lifestyle_numeric_data.infringed == 1.0]
non_infringed_lifestyle = lifestyle_numeric_data[lifestyle_numeric_data.infringed == 0.0]
```

```python
correlation_lifestyle = lifestyle_numeric_data.corr()
plt.figure()
sns.heatmap(correlation_lifestyle,
    xticklabels=lifestyle_numeric_data.columns,
    yticklabels=lifestyle_numeric_data.columns,
)
plt.title('Correlation between lifestyle and infringement')
plt.show()
```

```python
print('Average of the information when it has infringed')
print(infringed_lifestyle.mean())
```

```python
print("Average of the information when it doesn't have infringed")
print(non_infringed_lifestyle.mean())
```

#### Analyze

Pode-se observar que :
- o tempo de trabalho das pessoas e a região onde vivem têm uma ligeira influência nas infrações que cometem. Estes valores mesmo assim são baixos porque existe muito mais dados de não-infrações do que de infrações.
- em média, as pessoas que cometeram infrações tém menos income que as pessoas que não cometeram.
- as pessoas que pedem mais empréstimos vivem na casa dos 30

### Analyze non-numerical data

```python
lifestyle_infringed = lifestyle_categorical_data[lifestyle_categorical_data.infringed == 1.0].drop(['infringed'], axis=1)
lifestyle_non_infringed = lifestyle_categorical_data[lifestyle_categorical_data.infringed == 0.0].drop(['infringed'], axis=1)
```

### Probability of each parameter of each lifestyle category when there are infringements

```python
# infringed
total_count = lifestyle_infringed.shape[0]
for column in lifestyle_infringed.columns:
    for value in lifestyle_infringed[column].unique():
        var_count = lifestyle_infringed[lifestyle_infringed[column] == value].shape[0]
        prob = var_count / total_count * 100
        print(f'{column} when value is {value} : {prob} %')
```

### Probability of each parameter of each lifestyle category when there are no infringements

```python
# non-infringed
total_count = lifestyle_non_infringed.shape[0]
for column in lifestyle_non_infringed.columns:
    for value in lifestyle_non_infringed[column].unique():
        var_count = lifestyle_non_infringed[lifestyle_non_infringed[column] == value].shape[0]
        prob = var_count / total_count * 100
        print(f'{column} when value is {value} : {prob} %')
```
