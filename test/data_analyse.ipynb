# Prepare data to send

```python
def attribute_random_values_to_column(dataset, column):
    # specially for the SK_ID_CURR column
    data = dataset.copy()
    # calculates unique values and their len
    unique_values = data[column].unique()
    len_unique_values = len(unique_values)
    # generates random codes for the unique values
    random_codes = np.arange(0, len_unique_values, 1)
    np.random.shuffle(random_codes)
    # maps the unique values into their codes
    mapped_values = dict(zip(unique_values, random_codes))
    #print(mapped_values)
    coded_column = data[column].map(mapped_values)
    # generates a new csv with the mapped values
    lookup_table_data = np.array([unique_values, random_codes]).T
    lookup_table = pd.DataFrame(data=lookup_table_data, columns=[column, 'code'])
    lookup_table.to_csv(os.path.join('dataset', f'{column}_lookup_table.csv'), index=False)
    return coded_column
```

```python
def remove_private_data(dataset, sensitive_columns):
    data = dataset.copy()
    data = data.drop(sensitive_columns, axis=1) # remove sensitive columns
    return data
```

```python
def prepare_data_to_send(data_path):
    # reads dataset
    data = pd.read_csv(data_path)

    # encodes column with id
    to_code_column = 'SK_ID_CURR'
    data[to_code_column] = attribute_random_values_to_column(data, to_code_column)

    # handles columns that contain private info about the clients
    columns_with_private_data = ['first_name','last_name', 'provided_mobilephone', 'provided_workphone', 'provided_homephone', 'provided_email']
    data_without_private_info = remove_private_data(data, columns_with_private_data)
    
    path_file2send = os.path.join('dataset', 'infringement_dataset_send.csv') # path to save the file

    data_without_private_info.to_csv(path_file2send, index=False ) # saves the dateset to be send

    return path_file2send
```

```python
path = prepare_data_to_send(infringement_path)
```
